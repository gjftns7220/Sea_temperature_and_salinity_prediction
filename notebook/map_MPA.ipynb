```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt


coordinates = [
    (-135.8405278, 53.0521111),
    (-134.9987222, 53.2724722),
    (-135.2846944, 53.6636667),
    (-135.89625, 53.655),
    (-136.5064167, 53.8713056),
    (-136.7925278, 53.8221111),
    (-136.9509722, 53.6673611),
    (-136.1666667, 53.2331111),
    (-135.8405278, 53.0521111) 
]
lons, lats = zip(*coordinates)
```

```python
import cartopy.crs as ccrs
import matplotlib.pyplot as plt


min_lon, max_lon = min(lons)-5, max(lons)+5
min_lat, max_lat = min(lats)-5, max(lats)+5


margin = 0.5


fig, ax = plt.subplots(figsize=(10, 10), subplot_kw={'projection': ccrs.PlateCarree()})


ax.coastlines(resolution='10m')
ax.stock_img()


ax.set_extent([min_lon - margin, max_lon + margin, min_lat - margin, max_lat + margin], crs=ccrs.PlateCarree())


ax.fill(lons, lats, color='lightblue', alpha=0.4)
ax.plot(lons, lats, color='blue')

gl = ax.gridlines(draw_labels=True, color='gray', alpha=0.5, linestyle='--')
gl.top_labels = gl.right_labels = False 
```
